```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

health = pd.read_csv("health.csv", index_col=0)
health
```

```python
le = health['le'].sort_values(ascending=False)
fig, ax = plt.subplots(1, 1)
ax.bar(range(len(le)), le, tick_label = le.index)
ax.set_xticklabels(le.index, rotation=90, fontsize="small")
ax.set_ylim(le.values.max() - 10, le.values.max())
```

```python
sk_le = health[["sk", "le"]].sort_values(by="sk", ascending=True)
fig2, ax2 = plt.subplots(1, 1)
ax2.scatter(sk_le["sk"].values, sk_le["le"].values)
ax.set_ylabel("Life Expectancy")
ax.set_xlabel("Smoking Rate")
```
